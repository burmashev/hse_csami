```python
import os
import numpy as np
import pylab as plt
import scipy.signal as sg
import sounddevice as sd
import pandas as pd

from time import sleep

from IPython.display import Audio, display
def play(x, fs):
    display(Audio(x, rate=fs))
```

```python
from sklearn.model_selection import train_test_split

# load data
data_dir = '/content/drive/MyDrive/НИС АДИС 2022/data_project'
x_global_train = np.load(os.path.join(data_dir, 'train.npz'))['x']
y_global_train = np.load(os.path.join(data_dir, 'train.npz'))['y']
x_global_test  = np.load(os.path.join(data_dir, 'test.npz'))['x']

# split train dataset into local train and test
x_train, x_test, y_train, y_test = train_test_split(x_global_train, y_global_train, test_size=0.15, random_state=42)
fs = 16000
```

```python
for k in np.random.choice(np.arange(len(x_train)), 3):
    print(int(y_train[k]))
    play(x_train[k], fs)
```

```python
fig, ax = plt.subplots(2, figsize=(4,4), dpi=150)
plt.tight_layout()

x = x_train[y_train==1][0]
print(x)

time = np.arange(len(x))/fs

segment = [0.4, 0.6]
ax[0].plot(time, x)
ax[0].set_xlabel('time, s')
[ax[0].axvline(b, color='r') for b in segment]
ax[1].plot(time, x)
ax[1].set_xlim(segment)
ax[1].set_xlabel('time, s')
play(x, fs)
```

```python
freq, x_spec_train = sg.welch(x_train, fs, axis=1, )
print(freq.shape, x_spec_train.shape)
plt.plot(freq, x_spec_train[y_train==1][1])
plt.semilogy(0, 4000)
```

```python
fig, ax = plt.subplots(1, figsize=(4,3), dpi=150)
for k in range(1, 3+1):
    specs = np.log(x_spec_train[y_train==k])
    plt.plot(freq, specs.mean(0), label=['One', 'Two', 'Three'][k-1])
    plt.fill_between(freq, specs.mean(0)-specs.std(0), specs.mean(0)+specs.std(0), alpha=0.1)
    
plt.legend()
plt.xlabel('Freq. Hz')
band = [(600, 900), (1000, 1300), (1400, 1700), (1800, 2100), (2200, 2400), (5000, 5300)]
[plt.axvline(l, color='r') and plt.axvline(r, color='r') for l, r in band]
```

```python
def extract_features(x, band):
    freq, spec = sg.welch(x, fs, axis=1)
    features = spec[:, ((freq>band[0]) & (freq<band[1]))].mean(1).reshape(-1, 1)
    return features
```

```python
f_train = []
f_test = []
f_global_test = []

for b in band:
  f_train_ = extract_features(x_train, b)
  f_test_ = extract_features(x_test, b)
  f_global_test_ = extract_features(x_global_test, b)
  if len(f_train) == 0:
    f_train = f_train_
    f_test = f_test_
    f_global_test = f_global_test_
  else:
    f_train = np.hstack([f_train, f_train_])
    f_test = np.hstack([f_test, f_test_])
    f_global_test = np.hstack([f_global_test, f_global_test_])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from catboost import CatBoostClassifier

from sklearn.model_selection import KFold, cross_val_score
cv = KFold(5, shuffle=True, random_state=42)

gs = CatBoostClassifier(random_state=42, iterations=300, learning_rate=0.2,
                        max_depth=3, max_leaves=8)
gs.fit(f_train, y_train)
```

```python
y_train_pred = gs.predict(f_train).astype(int)
y_test_pred = gs.predict(f_test).astype(int)
y_global_test_pred = gs.predict(f_global_test).astype(int)

accuracy_score(y_train, y_train_pred), accuracy_score(y_test, y_test_pred)
```
