```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('diamonds.csv')
data
```

```python
from sklearn.cluster import KMeans

def clasters(data, n_clusters):
    criterions = []
    labels = []

    for i in range(10):
        kmeans = KMeans(n_clusters=n_clusters, init='random', n_init=1)
        kmeans.fit(data)
        criterions.append(kmeans.inertia_)
        labels.append(kmeans.labels_)
        
#     print(criterions)
#     print()
#     print(labels)

    opt_ind = np.argmin(criterions)
    opt_criterion = criterions[opt_ind]
    opt_label = labels[opt_ind]
    
    return criterions, labels, opt_criterion, opt_label
```

```python
num = ['carat', 'price', 'depth', 'x', 'y', 'z']
```

```python
data_norm = (data[num]-data[num].mean()) / data[num].std()
```

```python
data_norm
```

```python
# clasters(data_norm, 5)
```

```python
criterions5, labels5, opt_criterion5, opt_label5 = clasters(data_norm, 5)
```

```python
data_norm.loc[:, 'label5'] = opt_label5
```

```python
data_norm
```

```python
d = data[num].copy()
```

```python
d.loc[:, 'label5'] = opt_label5
d
```

```python
cl5 = d.groupby('label5').mean()
cl5
```

```python
tmp5 = d.mean()
tmp5
```

```python
new_cl5 = cl5.copy()
new_cl5.loc['Grand mean', :] = tmp5
new_cl5
```

```python
delta = (cl5 - tmp5) / tmp5 * 100
```

```python
delta[num]
```

```python
cnt = d.groupby('label5').size()
cnt
```

```python
delta.loc[:, 'cnt'] = cnt
```

```python
delta5 = delta[['carat', 'depth', 'price', 'x', 'y', 'z', 'cnt']]
```
